```python
y=data.pH
y.head()
```

```python
#Visualization
plt.plot(x,y,'o')
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(x,y)
print("Model coefficient: %.5f, and intercept: %.5f" %(model.coef_[0],model.intercept_))
```

```python
for h in model.coef_:
    print "coefficient: %f" %h
```
