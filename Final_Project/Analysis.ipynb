## First Goal

The dataset to be used to buld the model consists all 64 games from the World Cup, and the difference in frequency for each of the 200 plus statistics between the two competitors. I'll use 63 of the data points to build my model, and then use the data associated with the Brazil vs Germany game to make a prediction.

The value to be predicted here is the goal difference, i.e. the difference between the goals scored by either team. In the case of Brazil vs Germany, this was 6. (Technically it should be -6, but Germany was stored as team1 in the dataframe. It doesn't make much of a difference, really.)

```python
df_match_info = pd.read_csv('wc_data.csv')
df_match_info = df_match_info.fillna(0)
df_match_info.head(5)
```

Checking to see that all the matches are accounted for. 

```python
len(df_match_info)
```

The first key step is to drop the row containing the information on the Brazil x Germany game.

```python
df_match_info1 = df_match_info.drop(df_match_info.index[12])
len(df_match_info1)
```

Now I take a look at all the statistics (plus the other nonstatistical columns like the teams who contested each game). 

```python
for col in df_match_info.columns:
    print(col)
```

At this step I do some manual feature selection. Metrics that are directly related to score of the game must be discarded.

```python
not_needed = ['assist_attempt_saved','assist_blocked_shot', 'result','assist_free_kick_won','assist_own_goal','assist_pass_lost',\
              'assist_penalty_won', 'att_assist_openplay', 'att_assist_setplay', 'att_hd_goal','att_ibox_goal', 'att_lf_goal',\
              'att_obox_goal', 'att_pen_goal','att_rf_goal','formation_place','mins_played','own_goals','team1','team2',\
              'game_started','goal_assist','goal_assist_deadball','goal_assist_intentional','goal_assist_openplay','goal_assist_setplay',\
              'goal_fastbreak','goals','goals_conceded','goals_conceded_ibox','goals_conceded_obox','goals_openplay','second_goal_assist',\
              'big_chance_missed','big_chance_scored','error_lead_to_goal','pen_goals_conceded','penalty_faced','penalty_save']

len(not_needed)
```

```python
df_match_info2 = df_match_info1.drop(df_match_info1[not_needed],axis=1)
df_match_info2.head()
```

```python
len(df_match_info2.columns)
```

Now I try to build a model to predict the score of Brazil vs Germany based on the 203 statistics left. Since I'm trying to predict a continuous variable, linear regression models would be best.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, ExtraTreesRegressor, BaggingRegressor

#from sklearn import cross_validation
#from sklearn.metrics import mean_squared_error
#from sklearn.preprocessing import PolynomialFeatures
#from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import train_test_split
```

Now I setup my x and y, training and testing variables, and get the Brazil vs Germany statistics prepared to be used in it in my model.

```python
X = df_match_info2.values
y = df_match_info1['goals'].values
x_train, x_test, y_train, y_test = train_test_split(X, y)

bra_ger = df_match_info[df_match_info['team2']==614][df_match_info['team1']==357]
bra_ger = bra_ger.drop(df_match_info1[not_needed],axis=1)
```

Initializing the models.

```python
models = {'LinearRegression': LinearRegression(),'RandomForestRegressor': RandomForestRegressor(max_depth=3),'DecisionTreeRegressor': DecisionTreeRegressor(max_depth=3),\
          'AdaRegressor': AdaBoostRegressor(),'ExtraTreesRegressor': ExtraTreesRegressor(max_depth=3),\
          'BaggingRegressor': BaggingRegressor()}
```

Now I run each model to see how well it does.

```python
for name, model in models.items():
    clf = model
    clf.fit(x_train, y_train)
    score = clf.score(x_test,y_test)
    print("%s had an accuracy score of %0.4f"% (name, score))
```

The Bagging Regressor performed best with an accuracy score of 0.4115. Let's see what score it predicts.

```python
clf2 = BaggingRegressor()
clf2.fit(x_train, y_train)
print("The average predicted score of the model is: ", clf2.predict(bra_ger)[0])
```

So Germany would win with a by a one goal difference according to the model. Not even close to what occured, but pretty much in line of what one would expect from a World Cup semifinal.

Now I'll try to improve the prediction by using a linear regression model more suited to the structure of my data set. The Lasso technique works works with with a large number of features and a small number of data points.

```python
from sklearn.linear_model import Lasso

#I do cross validation by choosing 10,000 evenly spaced alphas between 0 and 100.

alphas = np.linspace(0,100, 10000)
scores = []

for alpha in alphas:
    
    lass = Lasso(alpha=alpha)
    lass.fit(x_train,y_train)
    scores.append(lass.score(x_test,y_test))
    
print(max(scores))
```

```python
best_alpha = alphas[scores.index(max(scores))]

lass2 = Lasso(alpha=alpha)
lass2.fit(x_train,y_train)
lass2.predict(bra_ger)[0]
```

## Second Goal

Brazil was without two of its best players in Neymar and Thiago Silva against Germany. What did they miss not not having them on the pitch.

```python
Image(filename='nt.jpg')
```

###### No Neymar, Brazil's Best Player

This question is much easier to answer for Neymar as attacking contribution in soccer is fairly easy to discern. Here's I'll make use of a subset of the metrics used by Ted Knutson, a wellknown mind in the soccer analytics community. The metrics are:

1. Goals
2. Shots
3. Shots on Target
4. Succesful Dribbles
5. Key Passes

The data set is a very small one that I constructed myself.

```python
dc = {'team/player': pd.Series(['Brazil','Neymar','Hulk','Oscar']),'shots': pd.Series([55,18,13,7]),'shots_on_target': pd.Series([36,9,5,5]),\
      'goals':pd.Series([11,4,0,2]),'key_passes': pd.Series([48,13,5,7]),'successful_dribbles':pd.Series([76,17,11,12])}

df = pd.DataFrame(dc)
df = df.set_index(df['team/player'].values)
df = df[['goals','key_passes','shots','shots_on_target','successful_dribbles']]
df.head()
```

The dataframe above shows the frequency of each metric in the first five games for Brazil, and Brazil's three best attacking players by overall statistical output: Neymar, Oscar, and Hulk. 

First I look at how much of Brazil's attacking output was due to Neymar.

```python
ax = df[:2].transpose().plot(kind='barh',title = "Brazil and Neymar's Attacking Output in the First Five Games")
ax.set_xlabel("total")
ax.set_ylabel("metrics")
```

Quite a lot. Let's see how he compared to Oscar and Hulk.

```python
ax2 = df[1:].transpose().plot(kind='barh',title = "Top 3 Brazilian Players by Attacking Output in the First Five Games")
ax2.set_xlabel("total")
ax2.set_ylabel("metrics")
```

He's ahead of them by quite some distance. This is what Brazil were missing going forward against Germany, and it was a twofold blow. Not only did the German defense have an easier time of things in his absence, but the knock-on effect also gave the Teutons less to worry about when they were attacking.

###### No Thiago Silva, Brazil's Best Defender

Trying to do something similar for Thiago Silva is much harder as defensive contribution is soccer is hard to quantify. Unlike attacking where the output is seen by what happens, being good defensively is often about what doesn't happen because of how well a player is performing.

Again I proceed by using some metrics identified by Ted Knutson as indicators of a good defender. But before I proceed, here's a table of Thiago Silva's ranking for each metric for Brazil during the World Cup compared to his teammates.

|  Metric | Interceptions | Tackles | Clearances | Shots Blocked | Fouls |
|:-------:|---------------|---------|------------|---------------|-------|
| Ranking | 8th           | 13th    | 1st        | 1st           | 15th  |

His rankings for interceptions and tackles are far from impressive, but those for the other three give an indication of how good he is. Fouls is different from the other metrics in that the less one commits, the better one is defensively. Thus, Silva's rank of fifteen is very good.

This data set is akin to the one in the Neymar section above.

```python
dc2 = {'team/player': pd.Series(['Brazil','Thiago Silva','David Luiz','David Luiz with Thiago Silva','David Luiz w/o Thiago Silva']),'interceptions': pd.Series([35,7,15,3,3]),'tackles': pd.Series([97,4,13,2.6,3]),\
      'clearances':pd.Series([151,45,37,7.2,0]),'shots_blocked': pd.Series([14,8,3,.6,0]),'fouls':pd.Series([82,4,12,2.4,5])}

df2 = pd.DataFrame(dc2)
df2 = df2.set_index(df2['team/player'].values)
df2 = df2[['interceptions','tackles','clearances','shots_blocked','fouls']]

df2.head()
```

A better way to approach the problem is to look at Thiago Silva along with his most frequency partner in the center of defense: David Luiz. The central defensive duo (or trio) is part of what is commonly referred to as the spine of the team in soccer. A bad one will certainly make a team lose games, and a good one plays its part in wins.

Althought David Luiz is a good defender, the consensus is that his has a wild streak that leads to him performing erratically. Another consensus is that Thiago Silva's presence beside him usually tames this behavior. 

Here's their defensive output along with Brazil's in the first five games.

```python
ax3 = df2[:3].transpose().plot(kind='barh',stacked = False,title = "The Defensive Output of Brazil, Thiago Silva, and David Luiz in the First Five Games")
ax3.set_xlabel("total")
ax3.set_ylabel("metrics")
```

It's clear that together they account for a lot of Brazil's output in that sector. Let me now switch gears to looking at David Luiz himself to see what Thiago Silva's absence meant. Here's David Luiz's defensive output when he plays beside Thiago Silva.

```python
ax4 = df2[3:4].transpose().plot(kind='barh',title="David Luiz\'s Defensive Output Playing with Thiago Silva")
ax4.set_xlabel("per game")
ax4.set_ylabel("metrics")
```

This in itself doesn't say much right now, but it will speak loudly enough later. The screenshot below gives a glimpse of how Luiz usually plays when partnered with Thiago Silva. He's positioned perfectly in the green circle, while Silva is in the blue circle. Positioning is one of the most important things in soccer that can't be measures, because if you're positioned right, that generally leads to an absence of activity.
