```python
#Imports

import json
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = 'http://data.huffingtonpost.com/2014/world-cup/matches/belgium-vs-usa-731822'
response = urlopen(url).read()
soup = BeautifulSoup(response)

links_html = soup.find_all("span", class_="matchup")

links = []

for link_html in links_html:
    a = link_html.find_all('a')
    for l in a:
        link = l.get('href')
        link = link.split('/')[-1]

    links.append(link)
    
links = list(set(links))
```

```python
def get_match_data(match):
    match_id = match.split('-')[-1]
    response = urlopen('http://data.huffingtonpost.com/2014/world-cup/matches/%s.json' % match_id).read()
    match_data = json.loads(response.decode())    
    return match_data

match_data = get_match_data(links[0])
match_data.keys()
```

```python
def get_match_names(match):    
        
    response = urlopen('http://data.huffingtonpost.com/2014/world-cup/matches/%s' % match) 
    soup = BeautifulSoup(response)

    data = {}

    data_script = soup.findAll("script")[1] 
    data_lines = data_script.text.split('\n')

    for line in data_lines[1:]:
        try:            
            line_data = line.split(' = ')
            name = line_data[0].split('.')[1]
            value = json.loads(line_data[1][:-1])
            data[name] = value
        except:
            print ("error parsing string: ", line)
            
    return data

names = get_match_names(links[55])
names['players']
```

Seems good so far.

```python
data = {}

for match in links:
    if match not in data:
        
        time.sleep(0.1)

        match_data = get_match_data(match)
        match_names = get_match_names(match)
        data[match] = {'data': match_data, 'names': match_names}

        print ("done")
        #print match
    else:
        print ("already processed")
        #print match
        
print (len(data.keys())) 
```
